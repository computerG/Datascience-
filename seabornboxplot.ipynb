```python
%matplotlib notebook
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
plt.rcParams.update({'figure.max_open_warning': 0})
```

```python
crash = sns.load_dataset('car_crashes')
```

```python
crash.head()
crash.min()
```

```python
plt.figure(figsize = (9, 9))

plt.subplot(3,2,1)
sns.boxplot(y=crash['total'])

plt.subplot(3,2,2)
sns.boxplot(y=crash['alcohol'])

plt.subplot(3,2,3)
sns.boxplot(y=crash['speeding'])



plt.subplot(3,2,4)
sns.boxplot(y=crash['not_distracted'])

plt.subplot(3,2,5)
sns.boxplot(y=crash['no_previous'])

plt.subplot(3,2,6)
sns.boxplot(y=crash['ins_premium'])

plt.subplot(3,2,7)
sns.boxplot(y=crash['ins_losses'])
```
